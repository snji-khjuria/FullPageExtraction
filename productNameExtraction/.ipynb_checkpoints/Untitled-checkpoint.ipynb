```python
#find all start, end pairs of particular key
def getAllStartEndPairs(document, key):
    keyLength = len(key)
    return [(i, i+keyLength) for i in range(len(document)) if document.startswith(key, i)]
 
    
#find seeds for one particular seed(at this stage call only getAllStartEndPairs())
def getEachSeedLocationsInPage(htmlPageContent, seed):
    return getAllStartEndPairs(htmlPageContent, seed)

def isInsideQuotes(document, seedsSet):
    totalLength = len(document)
    for s in seedsSet:
        allSeedPositions = getAllStartEndPairs(document, s)
        for (start, end) in allSeedPositions:
            prevLoc = start-1
            nextLoc = end
            if prevLoc>=0 and nextLoc<totalLength:
                prevQuot = document[prevLoc]
                nextQuot  = document[nextLoc]
                print(prevQuot + " " + nextQuot)
                if prevQuot=="\"" and nextQuot=="\"":
                    return True
    return False
            
s = ["sanjeev", "kumar"]
data = "my name is value = \"sanjeev\" kumar and yours is"
# isInsideQuotes(data, s)


def getLastWordWithoutEqualSign(leftContext):
    words = leftContext.strip().split(" ")
    totalWords = len(words)
    if len(words)<2:
        return ""
    else:
        if words[totalWords-1]=="=":
            return words[totalWords-2]
        else:
            return ""


def getSpecialWords(document, seedsSet):
    totalLength = len(document)
    specialWords = []
    for s in seedsSet:
        allSeedPositions = getAllStartEndPairs(document, s)
        for (start, end) in allSeedPositions:
            prevLoc = start-1
            nextLoc = end
            if prevLoc>=0 and nextLoc<totalLength:
                prevQuot = document[prevLoc]
                nextQuot  = document[nextLoc]
                print(prevQuot + " " + nextQuot)
                if prevQuot=="\"" and nextQuot=="\"":
                    leftContext = document[max(0, prevLoc-100):prevLoc]
                    lastWord = getLastWordWithoutEqualSign(leftContext)
                    if len(lastWord)>0:
                        specialWords.append(lastWord)
    return specialWords

words = getSpecialWords(data, s)



import re
def doProcessingWithoutClass(s, words):
    i = "class[\s]*=[\s]*"
    o       = "class="
    s = re.sub(i, o, s)
    i = "([^class])=\"[^\"]*\""
    o = "\\1=\"\""
    return re.sub(i, o, s)



def getInquotesRemover(wordsStr):
    return ("((" + wordsStr + "))=\"[^\"]*\"", "\\1=\"\"")


def getWordsAsString(words):
    out = words[0]
    if len(words)==1:
        return out
    else:
        for index in range(1, len(words)):
            out+="|"+words[index]
    return out

def getWhiteSpaceRemover(wordsStr):
    return ("(?!" + wordsStr+")[\s]*=[\s]*", "\\1=")

test = "my name is sanjeev = \"kumar\" = verma and name = \"kumar\" and name=kumar and pasha=\"value\" and yours"
print(re.sub("name|sanjeev", "Java", test))
wordsStr = getWordsAsString(["name", "sanjeev"])
(whitespaceRegExp, whiteSpaceReplacement) = getWhiteSpaceRemover(wordsStr)
(inQuotesMatch, inQuotesRemover) = getInquotesRemover(wordsStr)
print(whitespaceRegExp)
print(test)
whites = re.sub(whitespaceRegExp, whiteSpaceReplacement, test)
print(whites)
print(inQuotesMatch)
nonwhites = re.sub(inQuotesMatch, inQuotesRemover, whites)
print(nonwhites)
```
